## Construction du dataframe final

### customer_unique_id

```python
# customer_unique_id unique
print(df_result['customer_unique_id'].nunique() / len(df_result.index) * 100)
```

```python
list_customer_unique_id = df_result['customer_unique_id'].unique().tolist()
df_final_result = pd.DataFrame(list_customer_unique_id)
df_final_result.rename(columns={ df_final_result.columns[0]: "customer_unique_id" }, inplace = True)
df_final_result.head(5)
```

```python
print(df_final_result['customer_unique_id'].nunique() / len(df_final_result.index) * 100)
```

### Nombre de commandes par client

```python
df_temp = df_result.groupby(['customer_unique_id', 'order_id']).nunique()
df_temp2 = df_temp.reset_index()["customer_unique_id"].value_counts()
df_temp2.columns = ['customer_unique_id', 'counts']
df_temp2.head(5)

dict_orders = {}
for element in df_temp2.iteritems():
    dict_orders[element[0]] = element[1]
    
df_final_result['number_orders'] = df_final_result['customer_unique_id'].map(dict_orders)
df_final_result.head(5)
```

```python
df_final_result['number_orders'].max()
```

### Nombre de produits achetés par client

```python
df_temp = df_result.groupby(['customer_unique_id', 'product_id']).nunique()
df_temp2 = df_temp.reset_index()["customer_unique_id"].value_counts()
df_temp2.columns = ['customer_unique_id', 'counts']
df_temp2.head(5)

dict_number_products_by_customer = {}
for element in df_temp2.iteritems():
    dict_number_products_by_customer[element[0]] = element[1]
    
df_final_result['number_products_by_customer'] = df_final_result['customer_unique_id'].map(dict_number_products_by_customer)
df_final_result.head(5)
```

```python
df_final_result['number_products_by_customer'].max()
```

### Catégorie la plus fréquente

```python
def most_frequent(List):
    return max(set(List), key = List.count)

most_sold_categories = df_result.groupby(['customer_unique_id'])['product_category_name_english']
dict_categories = {}
for g, data in most_sold_categories: 
    #print(g)
    cats = data.tolist()
    cat = most_frequent(cats)
    #print(cat)
    dict_categories[g] = cat
df_final_result['most_category'] = df_final_result['customer_unique_id'].map(dict_categories)
df_final_result.head(5)
```

### Montant moyen

```python
average_amount = df_result.groupby(['customer_unique_id'])['price'].mean()
df_final_result['average_amount'] = df_final_result['customer_unique_id'].map(average_amount)
df_final_result.head(5)
```

### Montant maximal

```python
max_amount = df_result.groupby(['customer_unique_id'])['price'].max()
df_final_result['max_amount'] = df_final_result['customer_unique_id'].map(max_amount)
df_final_result.head(5)
```

```python
df_final_result['max_amount'].max()
```

### Nombre moyen de produits par commande

```python
df_final_result['average_product_by_order'] = df_final_result['number_products_by_customer'] / df_final_result['number_orders']
df_final_result.head(5)
```

```python
df_final_result['average_product_by_order'].max()
```

### Commande la plus récente

```python
from datetime import datetime
from datetime import date
from dateutil import parser

most_recent_order = df_result.groupby(['customer_unique_id'])['order_purchase_timestamp'].max()

d0 = date(2022, 4, 1)

most_recent_order_days = {}
most_recent_order_date = {}
for key in most_recent_order.keys():
    _date_time = most_recent_order[key]
    d1 = parser.parse(_date_time).date()
    delta = d0 - d1
    most_recent_order_days[key] = delta.days
    most_recent_order_date[key] = d1

df_final_result['most_recent_order_days'] = df_final_result['customer_unique_id'].map(most_recent_order_days)
df_final_result['last_date'] = df_final_result['customer_unique_id'].map(most_recent_order_date)

df_final_result.head(5)
```
